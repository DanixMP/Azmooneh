```python
# soft_skills_swot.py

questions = [
    # نقاط قوت
    {"q": "نقاط قوت شما چیست؟ توانایی‌ها یا مهارت‌هایی که شما را از دیگران متمایز می‌کند؟", "type": "strength"},
    {"q": "چه مهارت‌های برجسته‌ای دارید؟", "type": "strength"},
    {"q": "نظر دیگران درباره نقاط قوت شما چیست؟", "type": "strength"},

    # نقاط ضعف
    {"q": "نقاط ضعف یا محدودیت‌های خود را لیست کنید.", "type": "weakness"},
    {"q": "آیا عادات نامطلوبی دارید؟ اگر بله، کدام‌ها؟", "type": "weakness"},
    {"q": "نظر دیگران درباره نقاط ضعف شما چیست؟", "type": "weakness"},

    # فرصت‌ها
    {"q": "چه فرصت‌هایی در محیط یا اطراف شما وجود دارد که می‌توانید از آن‌ها برای رشد استفاده کنید؟", "type": "opportunity"},
    {"q": "چه مهارت‌های جدیدی می‌توانید یاد بگیرید تا مزیت رقابتی پیدا کنید؟", "type": "opportunity"},

    # تهدیدها
    {"q": "چه تهدیدهایی ممکن است مانع پیشرفت شما شوند؟", "type": "threat"},
    {"q": "آیا ویژگی‌های شخصی شما ممکن است رسیدن به اهداف را محدود کنند؟", "type": "threat"},
    {"q": "چه موانع خارجی (مثل رقابت یا تغییرات تکنولوژی) ممکن است شما را تهدید کنند؟", "type": "threat"}
]

# ذخیره پاسخ‌ها بر اساس دسته‌بندی
swot_answers = {
    "strength": [],
    "weakness": [],
    "opportunity": [],
    "threat": []
}

print("تحلیل SWOT شخصی")
print("به هر سوال پاسخ دهید و پاسخ خود را تایپ کنید.")
print("-" * 50)

for i, q in enumerate(questions):
    print(f"\nسوال {i+1} ({q['type'].capitalize()}): {q['q']}")
    ans = input("پاسخ شما: ").strip()
    swot_answers[q['type']].append(ans)

# نمایش تحلیل SWOT
print("\n" + "="*50)
print("نتایج تحلیل SWOT شما:")
for category in ["strength", "weakness", "opportunity", "threat"]:
    print(f"\n{category.capitalize()}:")
    for idx, answer in enumerate(swot_answers[category], 1):
        print(f"{idx}. {answer}")
print("="*50)
print("می‌توانید از این تحلیل برای برنامه‌ریزی و رشد شخصی استفاده کنید.")
```
